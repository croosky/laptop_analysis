## Разделение столбцов и приведение типов

```python
df
```

```python
for column in df.columns:
    print(f"{column}")
    print(df[column].unique())
    print("\n")
```

```python
df[df['Inches'].str.contains(r'\?')]
```

```python
df = df[df['Inches'] != '?']
```

```python
df['Inches'] = df['Inches'].astype(float) 
```

### ScreenResolution

```python
df['ScreenResolution'].unique()
```

```python
df['IPS'] = df['ScreenResolution'].str.contains(r'\bIPS\b', case=False)
```

```python
df['Touchscreen'] = df['ScreenResolution'].str.contains(r'\bTouchscreen\b', case=False)
```

```python
df['Screen Quality'] = df['ScreenResolution'].str.extract(r'(Full HD|4K Ultra HD|Retina Display|Quad HD+)', expand=False)
df['Screen Quality'] = df['Screen Quality'].fillna('None')
```

```python
df['ScreenResolution'] = df['ScreenResolution'].str.extract(r'(\d+x\d+)', expand=False)
```

```python
df
```

### CPu

```python
df['Cpu'].unique()
```

```python
df['CpuFrequency'] = df['Cpu'].str.extract(r'(\d+\.\d+GHz)', expand=False) #находим цифры d+   .   цирфы d+ GHz
df['CpuFrequency'] = df['CpuFrequency'].str.replace('GHz','').astype(float)
df['CpuModel'] = df['Cpu'].str.replace(r'\s*\d+\.\d+GHz', '', regex=True) #удаляем любые символы пробела (в том числе табы) s* и число герц
```

```python
df
```

```python
df.drop('Cpu', axis=1)
```

### Ram

```python
df['Ram'] = df['Ram'].str[:-2].astype(int)
```

```python
df
```

### Memory

```python
df['Memory'].unique()
```

```python
types = ['SSD', 'HDD', 'Flash Storage', 'Hybrid']
df['MemoryType'] = df['Memory'].apply(lambda x: next((t for t in types if t in x), None))
```

```python
df
```

```python
df['MemoryType'].unique()
```

```python
df[df['MemoryType'].isna()]
```

```python
df = df[df['Memory'] != '?']
```

```python
df['MemorySize'] = df['Memory'].apply(lambda x: x.split()[0] if 'GB' in x or 'TB' in x else None)
```

```python
df
```

```python
df['MemorySize'].unique()
```

```python
df['MemorySize'] = df['MemorySize'].apply(
    lambda x: int(x[:-2]) if x[-2:] == 'GB' else int(float((x[:-2]))) * 1024 if x[-2:]=='TB' else None
)
```

```python
df['AdditionalMemory'] = df['Memory'].apply(lambda x: x.split('+')[1].strip() if '+' in x else None)
```

```python
df['AdditionalMemory'].unique()
```

```python
df['AdditionalMemorySize'].unique()
```

```python
df['AdditionalMemoryType'].unique()
```

```python
types = ['SSD', 'HDD', 'Hybrid']
df['AdditionalMemoryType'] = df['AdditionalMemory'].apply(
    lambda x: next((t for t in types if t in x), None) if x else None
)
```

```python
df['AdditionalMemorySize'] = df['AdditionalMemory'].apply(
    lambda x: x.split()[0] if x and ('GB' in x or 'TB' in x) else None
)
```

```python
df=df.drop(['AdditionalMemoryType','AdditionalMemorySize'], axis=1)
```

```python
df
```

### gpu

```python
df['Gpu'].unique()
```

```python
df['GpuBrand'] = df['Gpu'].apply(lambda x: x.split()[0])
df['GpuModel'] = df['Gpu'].apply(lambda x: ' '.join(x.split()[1:]))
```

```python
df
```

### weight, price

```python
df['Weight']=df['Weight'].str[:-2]
```

```python
df['Weight'].unique()
```

```python
df = df[df['Weight'].notna() & (df['Weight'] != 0.0002)]
```

```python
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')
```

```python
df['Price'] = df['Price'].apply(lambda x: int(x))
```

### удаляем ненужные столбцы

```python
df=df.drop(['Cpu','Memory','Gpu'], axis=1)
```

```python
order=['Company','TypeName', 'Inches', 'ScreenResolution','IPS','Touchscreen','Screen Quality', 'Ram',
'CpuFrequency',	'CpuModel',	'MemoryType'	,'MemorySize'	,'AdditionalMemory', 'GpuBrand',	'GpuModel', 'OpSys',	'Weight',	'Price']
```

```python
df = df[order]
```

```python
df
```

```python
df = df.reset_index(drop=True)
```

# EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
```

```python
df.describe()
```

```python
categorical_columns = ['Company', 'TypeName', 'ScreenResolution', 'Screen Quality', 
                       'CpuModel', 'MemoryType', 'GpuBrand', 'GpuModel', 'OpSys']
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))

for i, j in enumerate(categorical_columns):
    row = i // 3
    col = i % 3
    df[j].value_counts().plot(kind='bar', ax=axes[row, col])
    axes[row, col].set_title(f'Распределение {j}')
    axes[row, col].set_ylabel('Частота')
    axes[row, col].tick_params(axis='x', rotation=45)
    print(df[j].value_counts())
    
plt.tight_layout()
plt.show()
```

```python
df.hist(bins=30, figsize=(15, 10))
plt.suptitle("Распределение числовых переменных", fontsize=16)
plt.show()
```

Боксплоты

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(df.select_dtypes(include=[np.number]).columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f'{col}')
plt.tight_layout()
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 8))
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Корреляционная матрица")
plt.show()
```

```python
sns.pairplot(df.select_dtypes(include=[np.number]), diag_kind='kde', plot_kws={'alpha':0.5})
plt.suptitle("Парные графики для числовых переменных", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
for i, col in enumerate(['Company', 'TypeName', 'ScreenResolution', 'Screen Quality', 
                        'MemoryType','AdditionalMemory', 'GpuBrand', 'OpSys'], 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=col, y='Price', data=df)  
    plt.title(f'Влияние {col} на цену')
    plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
